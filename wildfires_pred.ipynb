```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
# https://www.kaggle.com/edhirif/predict-the-causes-of-wildfires-using-python
```

```python
conn = sqlite3.connect('Data/FPA_FOD_20170508.sqlite')
```

```python
df = pd.read_sql_query(
    """SELECT 
        FIRE_YEAR,
        DISCOVERY_DATE,
        DISCOVERY_DOY,
        STAT_CAUSE_CODE,
        STAT_CAUSE_DESCR,
        FIRE_SIZE,
        STATE,
        LATITUDE,
        LONGITUDE
        
        FROM Fires""", conn)
df.head()
```

```python
df.info()
```

```python
df['DATE'] = pd.to_datetime(df['DISCOVERY_DATE'] - pd.Timestamp(0).to_julian_date(), unit='D')
df['DATE'].head()
```

```python
df['FIRE_MONTH'] = pd.DatetimeIndex(df['DATE']).month
df['FIRE_MONTH'].head()
```

```python
df['DAY_OF_WEEK'] = df['DATE'].dt.weekday
df['DAY_OF_WEEK'].head()
```

```python
df['DAY_OF_WEEK'].value_counts()
```

```python
df['DAY_NAME'] = df['DATE'].dt.day_name()
df['DAY_NAME'].head()
```

```python
df['count'] = 1
pvweek = pd.pivot_table(data=df, values='count', index='DAY_NAME', columns='STAT_CAUSE_DESCR', aggfunc=np.sum )
pvweek
```

```python
fires_day = df[['DAY_NAME','STAT_CAUSE_DESCR','count']]
pvday = fires_day.groupby(by=['DAY_NAME', 'STAT_CAUSE_DESCR']).count()
pvday
```

```python
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
data = df[['FIRE_YEAR', 'DISCOVERY_DATE', 'STAT_CAUSE_CODE',
       'STAT_CAUSE_DESCR', 'FIRE_SIZE', 'STATE', 'LATITUDE', 'LONGITUDE', 'FIRE_MONTH', 'DAY_OF_WEEK']]
data.head()
```

```python
data = data[data['STAT_CAUSE_DESCR']!='Missing/Undefined']
data.head()
```

```python
le = preprocessing.LabelEncoder()
data['STATE'] = le.fit_transform(data['STATE'])
data.head()
```

```python
X = data.drop(labels=['STAT_CAUSE_CODE', 'STAT_CAUSE_DESCR',],axis=1)
y = data['STAT_CAUSE_DESCR']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)
```

```python
rfc = RandomForestClassifier(n_estimators=50)
```

```python
rfc.fit(X_train,y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,rfc_pred))
print('\n')
print(classification_report(y_test,rfc_pred))
```

```python
forest = RandomForestClassifier(n_estimators=100)
```

```python
forest.fit(X_train,y_train)
```

```python
forest_pred = forest.predict(X_test)
```

```python
print(confusion_matrix(y_test,forest_pred))
print('\n')
print(classification_report(y_test,forest_pred))
```
