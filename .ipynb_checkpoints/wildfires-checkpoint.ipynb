```python
#https://www.fs.usda.gov/rds/archive/products/RDS-2013-0009.4/_metadata_RDS-2013-0009.4.html#Spatial_Reference_Information
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#Need breakdown of fires started by humans and the size of fire
# Questons: What predicts the size of the fire? Location, duration, cause?
```

```python
conn = sqlite3.connect('Data/FPA_FOD_20170508.sqlite')
```

```python
c = conn.cursor()
```

```python
cursor = conn.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = [
        v[0] for v in cursor.fetchall()
        if v[0] != "sqlite_sequence"
    ]
cursor.close()
print(tables)
```

```python
df = pd.read_sql_query("select * from Fires", conn)
df.drop('Shape',axis=1, inplace=True)
df.head()
```

```python
sns.set()
plt.figure(figsize=(17,8))
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='normal',
    fontsize='medium'  
)
sns.countplot(x='STAT_CAUSE_DESCR', data=df, palette='autumn', order=df['STAT_CAUSE_DESCR'].value_counts().index)
plt.xlabel('Cause')
plt.ylabel('Number of fires')
plt.title('Number of fires in the United States (1992-2015) per cause')
```

```python
df_years = pd.DataFrame(df['FIRE_YEAR'].value_counts())
df_years.rename(columns={"FIRE_YEAR": "fire_count"}, inplace=True)
df_years
```

```python
sns.heatmap(df_years,cmap='icefire')bar
```

```python
fires = df[['FIRE_YEAR', 'STAT_CAUSE_DESCR']]
fires['count'] = 1
fires.rename(columns={"STAT_CAUSE_DESCR": "Cause"}, inplace=True)
fires
```

```python
pvfires = fires.groupby(by=['FIRE_YEAR', 'Cause']).count()['count'].unstack()
pvfires
```

```python
plt.figure(figsize=(17,8))
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='normal',
    fontsize='medium'  
)
sns.heatmap(data=pvfires, cmap='icefire')
plt.xlabel('Cause')
plt.ylabel('Year')
plt.title('Cause by Year')
```

```python
import datetime
df['FIRE_DATE'] = pd.to_datetime(df['DISCOVERY_DATE'] - pd.Timestamp(0).to_julian_date(), unit='D')       
```

```python
df['FIRE_DATE'].head()
```

```python
df['FIRE_MONTH'] = df['FIRE_DATE'].apply(lambda x: x.strftime('%B'))
```

```python
df['FIRE_MONTH'].head()
```

```python
months = df[['FIRE_YEAR', 'FIRE_MONTH']]
months['count'] = 1
months.head()
```

```python
pvmonths = months.groupby(by=['FIRE_YEAR', 'FIRE_MONTH']).count()['count'].unstack()
pvmonths
```

```python
data_pivot = months.pivot_table(values='count', index='FIRE_YEAR', columns='FIRE_MONTH', aggfunc=np.sum)
data_pivot
```

```python
pvmonths = pvmonths.loc[:,['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August','September', 'October', 'November', 'December']]
```

```python
plt.figure(figsize=(17,8))
sns.heatmap(data=pvmonths, linewidths=0.05, cmap='Oranges', fmt='1.0f', annot=True)
plt.title('Heatmap of number of fires in the US in each month of the year')
plt.xlabel('Month')
plt.ylabel('Year')
```

```python
pivot = fires.groupby(by=['FIRE_YEAR', 'Cause']).count()
pivot
```

```python
plt.figure(figsize=(17,10))
sns.lineplot(x='FIRE_YEAR',y='count',data=pivot.reset_index(),hue='Cause',palette='tab10')
plt.xlabel('Year')
plt.ylabel('Number of fires')
plt.title('Number of fires per year')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
